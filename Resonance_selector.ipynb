```python
#function for remove the GC envelope and find the periodicity of the resonance pattern

from scipy import signal
from scipy.signal import argrelextrema
import scipy.optimize
import scipy.fftpack
import math
#plt.rcParams['figure.dpi'] = 300

def RingAnalysis(Measured_Data,spectreName,DEBUG_MODE):
    
    lambda_center=1.56 # µm
    time_step = np.mean(np.diff(Measured_Data[:, 0]))
    Measured_Data_Power_Lin = 10 ** (Measured_Data[:, 1] / 10)
    sig_fft = scipy.fftpack.fft(Measured_Data_Power_Lin)
    power = np.abs(sig_fft) ** 2
    sample_freq = scipy.fftpack.fftfreq(Measured_Data_Power_Lin.shape[0], d=time_step)
    maxLocalFFT = argrelextrema(power, np.greater, order=4)
    peak_freq = np.median(np.diff(sample_freq[maxLocalFFT[0][0:8]]))
    FSR=1.0/peak_freq
    indexParPeriode = (1 / peak_freq) / np.nanmedian(np.diff(Measured_Data[:, 0])) + 1
    
    ## Detection des minimum (peak)
    coeffEnvelop = np.polyfit(Measured_Data[:, 0], Measured_Data[:, 1], 3)
    dataEnvelop=np.polyval(coeffEnvelop, Measured_Data[:, 0])
    # indexMinLocal_TH, infoMin = signal.find_peaks(-np.subtract(Measured_Data[:,1],dataEnvelop), distance=int(0.5*indexParPeriode), prominence=0.3*np.abs(np.nanmedian(np.subtract(Measured_Data[:,1],dataEnvelop))-np.nanmin(np.subtract(Measured_Data[:,1],dataEnvelop))))
    # indexMinLocal_TH, infoMin = signal.find_peaks(-np.subtract(Measured_Data[:,1],dataEnvelop), distance=len(Measured_Data[:, 0])*(lambda_center**2/ng/2/math.pi/R)/(Measured_Data[len(Measured_Data[:, 0])-1, 0]-Measured_Data[0, 0]), prominence=(1.6,None))
    indexMinLocal_TH, infoMin = signal.find_peaks(-np.subtract(Measured_Data[:,1],dataEnvelop), distance=300, prominence=(1.6,None))
    
    ## Detection de l'envelope du signal
    b, a = scipy.signal.butter(3, 0.05)
    filtered = scipy.signal.filtfilt(b, a, Measured_Data[:, 1])
    coeffEnvelop = np.polyfit(Measured_Data[:, 0], filtered, 3)
    dataEnvelopFiltre = np.polyval(coeffEnvelop, Measured_Data[:, 0])
    indexLocalMaxPower, _ =signal.find_peaks(np.subtract(filtered,dataEnvelopFiltre), distance=int(0.3*indexParPeriode))
    indexLocalMinPower, _ =signal.find_peaks(np.subtract(dataEnvelopFiltre,filtered), distance=int(0.05*indexParPeriode),prominence=(None, 0.20*np.abs(np.nanmedian(np.subtract(Measured_Data[:,1],dataEnvelop))-np.nanmin(np.subtract(Measured_Data[:,1],dataEnvelop)))))
    indexLocalMaxPower=indexLocalMaxPower.astype(int)
    indexLocalMinPower=indexLocalMinPower.astype(int)
    DataEnvelopMedianeBruit = np.add(np.polyval(np.polyfit(Measured_Data[indexLocalMaxPower, 0], Measured_Data[indexLocalMaxPower, 1], 5), Measured_Data[:, 0]),np.polyval(np.polyfit(Measured_Data[indexLocalMinPower, 0], Measured_Data[indexLocalMinPower, 1], 5), Measured_Data[:, 0]))/2.0
    
    ## Subtract envelope and linearize
    const=0
    dataFlat=np.subtract(Measured_Data[:, 1],DataEnvelopMedianeBruit + const)
    dataFlatLin= 10 ** (dataFlat / 10)
    
    return(indexMinLocal_TH, infoMin, DataEnvelopMedianeBruit, dataFlatLin, FSR)

print('done!')
```
